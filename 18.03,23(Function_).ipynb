#**Factorial of n numbers**

```python

#factorial of n numbers
def factorial(n):
  if n==0:
    return 1
  else:
      return (n*factorial(n-1))
def fact():
  n=int(input("Enter the number:"))
  print(f"The factorial of {n} number is {factorial(n)}")
```

#**Sum_Of_Digit**

```python
# sum of the digit
def sod(n):
    #n=int(input("Enter the number:"))
    sum=0
    while n>0:
        remainder=n%10
        sum=sum+remainder
        n//=10
    return sum
def sum_of_digit():
    n=int(input("Enter the number:"))
    digit_sum=sod(n)
    print("The sum of digit =",digit_sum)
      
```

# **Fibonacci series**

```python
# Fibonacci series

def fibonacci():
  n=int(input("Enter the number:"))
  if n<=0:
    print("Invalid input!")
  elif n==1:
    print("fibonacci series of",n,"terms.")
    print(0)
  else:
    a,b=0,1
    fib=[a,b]
    for i in range(2,n):
        c=a+b
        a=b
        b=c
        fib.append(c)
    print("Fibonacci series upto",n,"terms.")
    print(fib)
    #fibonacci(10)
```

# **Prime Numbers**

```python
def prime():
  n=int(input("Enter the number:"))
  for i in range(2,n):
    if(n%i==0):
      print("Not prime number.")
    else:
      print("Prime number")  
```

#**Find the GCD(Greatest Comman divisor)of a number**

```python
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

def find_gcd():
    n = int(input("Enter the number of values: "))
    values = []
    for i in range(n):
        value = int(input(f"Enter value {i+1}: "))
        values.append(value)
    result = values[0]
    for i in range(1, n):
        result = gcd(result, values[i])
    print("The GCD is:", result)
```

#**Armstrong number**

```python
def armstrong():
  n=int(input("Enter the numbers:"))
  sum=0
  temp=n
  while temp>0:
    remainder=temp%10
    sum=sum+remainder*remainder*remainder
    temp//=10
  if n==sum:
    print(f"{n} is an Armstrong numbers")
  else:
    print(f"{n} is not an armstrong number")
#  def armstrong():
 #   n=int(input("Enter a number :"))



```

#**Strong Number**

```python
def factorial(num):
    if num == 0:
        return 1
    else:
        return num * factorial(num - 1)

def strong(num):
    sum = 0
    temp = num
    while temp > 0:
        digit = temp % 10
        sum += factorial(digit)
        temp //= 10
    return sum == num

def main():
    num = int(input("Enter a number: "))
    if strong(num):
        print(num, "is a strong number")
    else:
        print(num, "is not a strong number")
```

#**Palindrome Number**

```python
def palindrome(n):
    # Convert the number to a string and reverse it
    reversed_n = str(n)[::-1]
    # Check if the reversed string is the same as the original string
    return str(n) == reversed_n

def numbers():
    # Get user input for a number
    n = int(input("Enter a number: "))
    if palindrome(n):
        print(f"{n} is a palindrome!")
    else:
        print(f"{n} is not a palindrome.")
```
